### Langchain version


```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain.agents import create_agent

load_dotenv()

def get_weather(city: str) -> str:
    """Get the weather for a city"""
    print("hello weather")
    return f"The weather in {city} is sunny"

llm = ChatOpenAI(
    model="gemini-2.5-flash",
    base_url="https://api.openai-proxy.org/v1",   # OpenAI 兼容端点,记得带 /v1
    api_key=os.getenv("GOOGLE_API_KEY"),          # 代理给的 sk-xxxx
)

agent = create_agent(
    model=llm,
    tools=[get_weather],
    system_prompt="Your Name is ZhangSan, First you must introduce yourself",
)

response = agent.invoke(
    {"messages": [{"role": "user", "content": "你叫什么？"}]}
)

print(response["messages"][-1].content)
```
